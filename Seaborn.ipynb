###**catplot()**:
stripplot(), swarmplot(), boxplot(), violinplot(), boxenplot(), pointplot(), barplot(), and countplot()

```python
df = sns.load_dataset('tips')
df.head()
```

```python
sns.catplot(x ='sex', y ='total_bill', hue = 'sex', jitter = 0.25, data =df )
plt.show()
```

**jitter**: to increase the randomness of the points on the plot

```python
sns.catplot(x ='sex', y ='total_bill', hue = 'smoker', kind = 'swarm', data =df )
plt.show()
```

```python
sns.catplot(x ='sex', y ='total_bill', hue = 'smoker', kind = 'violin', palette =['g','b'], data =df )
plt.show()
```

```python
sns.catplot(x = 'sex',y = 'total_bill', hue = 'smoker', kind = 'box', data = df)
```
